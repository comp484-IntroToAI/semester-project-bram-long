```python
import pandas as pd
import requests
import gzip
import io

def get_recent_data(buoy):
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct']
    final_df = pd.DataFrame()

    for month in months:
        # Construct the URL for the gzipped text file
        url = f"https://www.ndbc.noaa.gov/view_text_file.php?filename={buoy}2024.txt.gz&dir=data/stdmet/{month}/"
        
        try:
            df = pd.read_csv(url, delim_whitespace=True)

            final_df = pd.concat([final_df, df], ignore_index=True)
        
        except requests.exceptions.HTTPError as e:
            print(f"HTTP Error {e.response.status_code}: {e.response.reason} - Skipping this month.")
        
        except Exception as e:
            print(f"An error occurred: {str(e)} - Skipping this month.")
            continue

    return final_df

# Example usage
buoy = 410492
result = get_recent_data(buoy)
print(result)

# https://www.ndbc.noaa.gov/view_text_file.php?filename=4104922024.txt.gz&dir=data/stdmet/Feb/
# https://www.ndbc.noaa.gov/view_text_file.php?filename=4104922024.txt.gz&dir=data/stdmet/Feb/
```

```python
def clean_data(data):
    df = data.copy()
    first_row = df.iloc[0]
    
    new_columns = []
    
    for i in range(len(df.columns)):
        col_name = df.columns[i]       # Get the column name
        col_value = first_row[i]             # Get the corresponding value from the first row
        new_col_name = f"{col_name} ({col_value})"  # Combine the column name and value in parentheses
        new_columns.append(new_col_name)  # Append the new column name to the list
    
    df.columns = new_columns
    
    df = df.drop(0).reset_index(drop=True)
    df = df.rename(columns = {"#YY (#yr)":"year","MM (mo)":"month","DD (dy)":"day","hh (hr)":"hour","mm (mn)":"minutes"})
    df = df[df['year'] != '#yr']
    df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour', 'minutes']])
    return df
station_41421_st_thomas_df = clean_data(final_df)
```
